想寫一個f(x)=exp^2x+7的微分，該怎麼寫呢?
$${f(x)}=e^{2x+7}$$ $${f'(x)}??$$

那f(x)=exp^2x+7積分，該怎麼寫呢?範圍(a,b)
$$\int_a^b e^{2x+7} dx$$

```python
x = sb("x") ##定義變數x
```

```python
y=cos(x)
```

```python
y.diff(x)
```

```python
y.integrate(x)
```

```python
x = symbols('x')
f = symbols('f', cls = Function)
f = x**4-1
print("f(x) =", f)
print("roots of f(x):", roots(f))
print("roots of f(2):", f.evalf(subs = {x:0}))
```

```python
x = np.linspace(-10,10,1000)
m = x**2
plt.plot(x,m)
```

```python
a = np.cos(x)
b = np.tan(x)
c = np.sin(x)
b[b>10]=np.inf
b[b<-10]=-np.inf
```

```python
plt.plot(x,a,label='$sin(x)$',color='green')
plt.plot(x,b,label='$tan(x)$',color='red')
plt.plot(x,c,label='$sin(x)$',color='black')
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,a,label='$sin(x)$',color='green')
plt.plot(x,b,label='$tan(x)$',color='red')
plt.plot(x,c,label='$cos(x)$',color='black')

plt.xlabel('period')
plt.ylabel('amplitude')
plt.title('PyPlot')
plt.xlim(0,10) 
plt.ylim(-2,2) 
plt.legend() 
```
